### Find out freashers and job holders

```python
def save():
    df.to_excel('current.xlsx', index=False)
```

```python
def filterFresher(row):
    return 'fresher' if row.experience == 0 else 'jobHolder'
#     return 1 if row.experience == 0 else 0
```

```python
df['fresher'] = df.apply(lambda row: filterFresher(row), axis=1)
df.head()
```

```python
def categorizeSubject(row):
    found = re.search(".*(?i)([bm] ?\.? ?sc|Computer|Science).*", row.degree) 
    return 0 if not found else 1
```

```python
df['engineer'] = df.apply(lambda row: categorizeSubject(row), axis=1)
df.head()
```

```python
def plot_distribution(data, features):
    i = 0
    plt.figure()
    col = 3
    row = int(np.ceil(len(features)/col))
    fig, ax = plt.subplots(row,col,figsize=(18,15))

    for feature in features:
        i += 1
        plt.subplot(row,col,i);
        sns.distplot(tuple(data[feature]))
        plt.xlabel(feature, fontsize=16)
        locs, labels = plt.xticks()
        plt.tick_params(axis='x', which='major', labelsize=12)
        plt.tick_params(axis='y', which='major', labelsize=12)
    plt.show();
    
plot_distribution(df, num_col)
```

```python
def plot_boxplot(data, features):
    i = 0
    plt.figure()
    col = 3
    row = int(np.ceil(len(features)/col))
    fig, ax = plt.subplots(row,col,figsize=(18,14))

    for feature in features:
        i += 1
        plt.subplot(row,col,i);
        sns.boxplot(tuple(data[feature]))
        plt.xlabel(feature, fontsize=16)
        locs, labels = plt.xticks()
        plt.tick_params(axis='x', which='major', labelsize=12)
        plt.tick_params(axis='y', which='major', labelsize=12)
    plt.show();
    
plot_boxplot(df, num_col)
```

```python
df.drop(["name", 'address','mobile','photo', "degree", "currentOffice",], axis=1, inplace=True)
cat_col = ["university", 'designation', "fresher", 'gender']
```

```python
df_encoded = pd.get_dummies(df, columns=cat_col, prefix=cat_col)

print("Now our dataset has {} columns.".format(df_encoded.shape[1]))
```

```python
df_encoded
```

```python
X = df_encoded.loc[:, df_encoded.columns!='matchingScore']
y = df_encoded.loc[:, df_encoded.columns=='matchingScore']
```

```python
from sklearn.preprocessing import Normalizer

scaler = Normalizer()
scaler.fit(X.values)
X_standered = scaler.transform(X.values)
X_scaled = pd.DataFrame(X_standered, columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.20, random_state=1234)

print("""
X_train has {} data points.
y_train has {} data points.
X_test has {} data points.
y_test has {} data points.
""".format(X_train.shape[0], y_train.shape[0], X_test.shape[0], y_test.shape[0]))
```

## 4. Data Modeling

```python
import sklearn
sklearn.__version__
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate
from sklearn.metrics import explained_variance_score, mean_absolute_error, r2_score
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor

cross_validate
```

```python
regressors = [AdaBoostRegressor(base_estimator=DecisionTreeRegressor(), random_state=1234),
              GradientBoostingRegressor(random_state=1234)]
val_results = dict()

for regessor in regressors:
    val_results[regessor.__class__.__name__] = cross_validate(regessor, X_train, y_train.values.ravel(), cv=5, scoring=('explained_variance', 'neg_mean_absolute_error', 'r2'), return_train_score=False)
    
val_results
```

```python
trts_result_table = pd.DataFrame(columns=['EV','MAE','R2'])

for regessor in regressors:
    regr = regessor.fit(X_train, y_train.values.ravel())
    y_pred_base = regr.predict(X_test)
       
    trts_result_table = trts_result_table.append({'EV':explained_variance_score(y_test, y_pred_base),
                                             'MAE':mean_absolute_error(y_test, y_pred_base),
                                             'R2':r2_score(y_test, y_pred_base)}, ignore_index=True)
    
trts_result_table.rename(index={0:'AdaBoostRegressor',
                                1:'GradientBoostingRegressor'}, inplace=True)

# display(Markdown("**Train-Test Performance**"))
trts_result_table
```
