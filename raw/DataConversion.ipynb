```python
import pandas as pd
```

```python
from os import listdir
from os.path import isfile, join

def getFilePaths(folder):
    f = [f for f in listdir(folder) if isfile(join(folder, f)) and f.endswith('.json')]
    return [folder + '/' + file for file in f]
```

```python
getFilePaths(folder='json')
```

```python
df = pd.DataFrame()
for f in getFilePaths(folder='json'):
    new_df = pd.read_json(f)
    new_df['designation'] = f.split('dataset_')[1].split('.')[0]
    df = df.append(new_df, ignore_index=True)
df.shape
```

```python
df.head()
```

```python
newNames = {'Name': 'name',
                   'Age': 'age',
                   'ApplicantLocation': 'address',
                   'Mobile': 'mobile',
                   'Photo': 'photo',
                   'University': 'university',
                   'Degree': 'degree',
                   'Exps': 'currentOffice',
                   'Exp': 'experience',
                   'designation': 'designation',
                   'Salary': 'salary',
                   'MatchingScore': 'matchingScore'}
df.rename(columns=newNames, inplace=True)

headers = list(newNames.values())
df = df.reindex(headers, axis=1)
df.head()
```

```python
df.currentOffice = df.currentOffice.str.split('</strong>').str.get(0).str.split('<strong>').str.get(1)
df.experience = df.experience.str.extract('(\d+)',expand=True).astype(int)
df.salary = df.salary.str.replace(',','').astype(int)
df.mobile = df.mobile.str.replace('<br>','')
df.head()
```
