# Populating Gender Column

```python
import requests, re
url = 'https://www.moderatecontent.com/api/v2?face=true&key=6f79661500c596cb62e5c9a908f3f4bd&url='
isMaleRegx = '^((md[\. :] ?)|(m[ou]ha?m+[ea]d))'  # checks for (Mohammad / Md.)
```

```python
def getGender(row):
    g = ''
    foundMD = re.findall(isMaleRegx, row['name'], re.IGNORECASE)
    if row['gender'] in ['male', 'female']:
        g = row['gender']
        print(row['gender'], '-', row['name'], '-', 'already exists...')
    elif foundMD:
        g = 'male'
        print(g, '-', row['name'])
    else:
        try:
            r = requests.get(url + row.photo)
            g = r.json()['faces'][0]['gender']
        except:
            g = ''
        print('XXX' if not g else g, '-', row['name'], '-', row.photo)
    return g
```

```python
# df.head(10).apply(lambda row: getGender(row), axis=1)
df['gender'] = df.apply(lambda row: getGender(row), axis=1)
```
